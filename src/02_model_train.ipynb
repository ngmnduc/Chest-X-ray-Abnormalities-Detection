```python
from ultralytics import YOLO
import os


if __name__ == '__main__':
    # Khởi tạo model phiên bản Small (v8s)
    model = YOLO('yolov8s.pt') 

    # Thiết lập đường dẫn tương đối phù hợp với cấu trúc repo Git
    data_path = os.path.join(os.getcwd(), '../data/yolo_dataset/dataset.yaml')

    # Huấn luyện mô hình đồng bộ với cấu hình v4_consensus đã chạy trên Colab
    results = model.train(
        data=data_path,
        name='v8_abnormality_detection_v4_consensus',
        
        epochs=50,
        imgsz=512,
        batch=32,
        patience=15,
        
        amp=True,        # Mixed precision để tăng tốc độ và giảm VRAM
        cache='ram',     # Cache data lên RAM để tăng tốc độ đọc từ epoch 2
        workers=4,       # Phù hợp với cấu hình máy Colab T4
        cls=2.0,         # Tăng trọng số loss phân loại để cải thiện Recall
        cos_lr=True,     # Sử dụng Cosine Annealing scheduler
        
        val=True,
        save=True,
        save_period=-1   # Chỉ lưu model tốt nhất và model ở epoch cuối cùng
    )

    print("Done training!")
```
