```python
from ultralytics import YOLO
import os

# Khởi tạo model
model = YOLO('yolov8n.pt') 
data_path = os.path.join(os.getcwd(), '../data/yolo_dataset/dataset.yaml')

# Huấn luyện mô hình với cấu hình nâng cao
results = model.train(
    # --- 1. Cấu hình cơ bản (Basic Configs) ---
    data=data_path,
    epochs=50,                  # Số vòng lặp qua toàn bộ dataset
    imgsz=512,                  # Kích thước ảnh đầu vào
    batch=16,                   # Số lượng ảnh đưa vào model mỗi bước học (tùy VRAM)
    name='v8_abnormality_detection', 
    
    # --- 2. Cấu hình phần cứng (Hardware & System) ---
    device=0,               # 'cpu' hoặc 0 (nếu có GPU)
    workers=16,                  # Số luồng CPU dùng để load dữ liệu (tăng lên 8, 16 nếu CPU mạnh)
    
    # --- 3. Tối ưu hóa & Quá trình học (Optimization & Learning) ---
    optimizer='auto',           # Trình tối ưu hóa ('SGD', 'Adam', 'AdamW', 'auto')
    patience=15,                # Early stopping: Dừng sớm nếu sau 15 epoch không cải thiện mAP (giúp tránh overfit)
    lr0=0.01,                   # Learning rate ban đầu (thường để 0.01 cho SGD, 0.001 cho Adam)
    lrf=0.01,                   # Tỉ lệ Learning rate cuối cùng so với ban đầu
    weight_decay=0.0005,        # Phạt các trọng số lớn để tránh Overfitting
    warmup_epochs=3.0,          # Số epoch chạy khởi động với learning rate thấp để model ổn định
    
    # --- 4. Tăng cường dữ liệu (Data Augmentation) ---
    # (YOLO tự động áp dụng nhưng bạn có thể chỉnh tay)
    degrees=0.0,                # Xoay ảnh (độ) - ảnh y tế thường ít xoay mạnh
    translate=0.1,              # Dịch chuyển ảnh
    scale=0.5,                  # Phóng to/thu nhỏ ảnh
    flipud=0.0,                 # Lật dọc ảnh (0.0 = không lật, ảnh X-quang không nên lật dọc)
    fliplr=0.5,                 # Lật ngang ảnh (50% xác suất lật ngang)
    mosaic=1.0,                 # Kỹ thuật ghép 4 ảnh thành 1 (rất hiệu quả với YOLO)
    
    # --- 5. Lưu trữ & Kiểm thử (Saving & Validation) ---
    val=True,                   # Đánh giá trên tập validation sau mỗi epoch
    save=True,                  # Lưu lại trọng số (weights)
    save_period=-1,             # Lưu checkpoint mỗi N epoch (-1 là chỉ lưu model cuối và tốt nhất)
    exist_ok=True               # Cho phép ghi đè lên thư mục cũ nếu trùng tên 'name'
)

print("Done training!")
```
